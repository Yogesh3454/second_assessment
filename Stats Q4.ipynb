```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate example dataset
np.random.seed(0)
```

```python
x = np.random.normal(0, 1, 100)  # Independent variable (random normal distribution)
```

```python
y = 2 * x + np.random.normal(0, 1, 100)  # Dependent variable with positive correlation
```

```python
# Create DataFrame
df = pd.DataFrame({'X': x, 'Y': y})
```

```python
# Calculate correlation coefficient
correlation_coefficient = df['X'].corr(df['Y'])
```

```python
print("Correlation coefficient:", correlation_coefficient)
```

```python
# Plot scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df)
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
